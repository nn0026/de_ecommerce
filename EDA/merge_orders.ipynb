# Gộp các file Order Excel 

```python
import pandas as pd
from pathlib import Path
import glob
```

```python
# thư mục chứa các file Excel
data_folder = Path('../raw_data/data')

# Lấy danh sách tất cả file xlsx
excel_files = sorted(data_folder.glob('*.xlsx'))
for f in excel_files:
    print(f' {f.name}')
```

```python
# Đọc và gộp tất cả file
dfs = []
for file in excel_files:
    df = pd.read_excel(file)
    df['source_file'] = file.name 
    dfs.append(df)
    print(f' Số dòng đã thêm: {len(df)}')    
# Gộp DataFrame
merged_df = pd.concat(dfs, ignore_index=True)
print(f'\n Tổng cộng {len(merged_df)} dòng')
```

```python
# Xem thông tin tổng quan
print('Các cột trong dữ liệu:')
merged_df.head()
```

```python
# Kiểm tra dữ liệu
merged_df.info()
```

```python
# Xuất ra file CSV
output_path = '../raw_data/orders_merged.csv'
merged_df.to_csv(output_path, index=False, encoding='utf-8-sig')
print(f'Đã lưu file: {output_path}')
print(f'Tổng số dòng: {len(merged_df)}')
```
