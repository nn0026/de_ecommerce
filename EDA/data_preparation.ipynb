#  Data Cleaning & Preparation 

## 1. Load Data

```python
import pandas as pd
import numpy as np
df = pd.read_csv('../raw_data/orders_merged.csv')
df.head()
```

## 2. Data Overview

```python
df.info()
```

```python
df.describe(include='all')
```

## 3. Missing Value Analysis

```python
missing = df.isna().sum().to_frame('missing_count')
missing['missing_pct'] = (missing['missing_count'] / len(df)) * 100
missing
```

## 4. Check Duplicate 

```python
dup_count = df.duplicated().sum()
dup_count
```

```python
df = df.drop_duplicates()
df.shape
```

## 5. Clean String Columns

```python
for col in df.select_dtypes(include='object'):
    df[col] = df[col].astype(str).str.strip().str.replace('\s+', ' ', regex=True)
df.head()
```

## 6. Date Conversion 

```python
date_cols = [col for col in df.columns if 'date' in col.lower() or 'timestamp' in col.lower()]
date_cols
```

```python
for col in date_cols:
    try:
        df[col] = pd.to_datetime(df[col], errors='coerce')
    except:
        pass
df[date_cols].head()
```

## 7. Numeric Standardization

```python
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols].describe()
```

## 8. Outlier Detection 

```python
outlier_report = []
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    LL = Q1 - 1.5 * IQR
    UL = Q3 + 1.5 * IQR
    low = (df[col] < LL).sum()
    high = (df[col] > UL).sum()
    outlier_report.append([col, LL, UL, low, high])

outlier_df = pd.DataFrame(outlier_report, columns=['col', 'lower_limit', 'upper_limit', 'low_count', 'high_count'])
outlier_df
```

## 9. Missing Value

```python
# Numeric columns 
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Object columns 
for col in df.select_dtypes(include='object'):
    df[col] = df[col].fillna('UNKNOWN')

# Datetime columns 
for col in date_cols:
    df[col] = df[col].fillna(df[col].min())

df.isna().sum().head()
```

## 10. Save Cleaned Data

```python
clean_path = '../raw_data/orders_cleaned.csv'
df.to_csv(clean_path, index=False)
clean_path
```
